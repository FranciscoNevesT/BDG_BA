```python
# https://dados.gov.br/dados/conjuntos-dados/cadastro-unico---familiaspessoas-por-faixas-de-renda-per-capita
```

```python
url_2021 = 'https://aplicacoes.mds.gov.br/sagi/servicos/misocial?fq=anomes_s:2021*&wt=csv&omitHeader=true&fq=cadunico_tot_fam_i:{0%20TO%20*]&q=*&fl=ibge:codigo_ibge,anomes:anomes_s,cadunico_tot_fam:cadunico_tot_fam_i,cadunico_tot_pes:cadunico_tot_pes_i,cadunico_tot_fam_rpc_ate_meio_sm:cadunico_tot_fam_rpc_ate_meio_sm_i,cadunico_tot_pes_rpc_ate_meio_sm:cadunico_tot_pes_rpc_ate_meio_sm_i,cadunico_tot_fam_pob:cadunico_tot_fam_pob_i,cadunico_tot_pes_pob:cadunico_tot_pes_pob_i,cadunico_tot_fam_ext_pob:cadunico_tot_fam_ext_pob_i,cadunico_tot_pes_ext_pob:cadunico_tot_pes_ext_pob_i,cadunico_tot_fam_pob_e_ext_pob:cadunico_tot_fam_pob_e_ext_pob_i,cadunico_tot_pes_pob_e_ext_pob:cadunico_tot_pes_pob_e_ext_pob_i&rows=100000000&sort=anomes_s%20desc,%20codigo_ibge%20asc'
url_2022 = 'https://aplicacoes.mds.gov.br/sagi/servicos/misocial?fq=anomes_s:2022*&wt=csv&omitHeader=true&fq=cadunico_tot_fam_i:{0%20TO%20*]&q=*&fl=ibge:codigo_ibge,anomes:anomes_s,cadunico_tot_fam:cadunico_tot_fam_i,cadunico_tot_pes:cadunico_tot_pes_i,cadunico_tot_fam_rpc_ate_meio_sm:cadunico_tot_fam_rpc_ate_meio_sm_i,cadunico_tot_pes_rpc_ate_meio_sm:cadunico_tot_pes_rpc_ate_meio_sm_i,cadunico_tot_fam_pob:cadunico_tot_fam_pob_i,cadunico_tot_pes_pob:cadunico_tot_pes_pob_i,cadunico_tot_fam_ext_pob:cadunico_tot_fam_ext_pob_i,cadunico_tot_pes_ext_pob:cadunico_tot_pes_ext_pob_i,cadunico_tot_fam_pob_e_ext_pob:cadunico_tot_fam_pob_e_ext_pob_i,cadunico_tot_pes_pob_e_ext_pob:cadunico_tot_pes_pob_e_ext_pob_i&rows=100000000&sort=anomes_s%20desc,%20codigo_ibge%20asc'
```

```python
cad_unico_2021 = pd.read_csv(url_2021)
cad_unico_2021
```

```python
cad_unico_2022 = pd.read_csv(url_2022)
cad_unico_2022
```

Separando coluna anomes

```python
cad_unico_2021['anomes'] = cad_unico_2021['anomes'].astype(str)
cad_unico_2021['ano'] = cad_unico_2021['anomes'].str[:4]
cad_unico_2021['mes'] = cad_unico_2021['anomes'].str[4:]
cad_unico_2021['ano'] = cad_unico_2021['ano'].astype(int)
cad_unico_2021['mes'] = cad_unico_2021['mes'].astype(int)
cad_unico_2021.drop('anomes', axis=1, inplace=True)
```

```python
cad_unico_2022['anomes'] = cad_unico_2022['anomes'].astype(str)
cad_unico_2022['ano'] = cad_unico_2022['anomes'].str[:4]
cad_unico_2022['mes'] = cad_unico_2022['anomes'].str[4:]
cad_unico_2022['ano'] = cad_unico_2022['ano'].astype(int)
cad_unico_2022['mes'] = cad_unico_2022['mes'].astype(int)
cad_unico_2022.drop('anomes', axis=1, inplace=True)
```

### Adicionar tabelas ao banco

```python
import sqlite3
```

```python
db_path = 'sql/eleicao.db'
connection = sqlite3.connect(db_path)
```

```python
table_name = 'cadastro_unico_2021'
cad_unico_2021.to_sql(table_name, connection, if_exists='replace', index=False)
```

```python
table_name = 'cadastro_unico_2022'
cad_unico_2022.to_sql(table_name, connection, if_exists='replace', index=False)
```
