### Adicionar tabela ao banco

```python
import sqlite3
```

```python
db_path = 'sql/eleicao.db' 
connection = sqlite3.connect(db_path)
```

```python
gdf.columns
```

```python
censo_agro = gdf[['OBJECTID', 'Shape_Leng', 'Shape_Area', 'num_estab_mun',
       'area_media_ha', 'media_pessoal_ocupado_estab', 'lavoura_temporaria',
       'lavoura_permanente', 'atividade_pecuaria', 'atividade_hort_flor', 
       'atividade_sem_mudas', 'atividade_florestal', 'atividade_pesca',
       'atividade_agric', 'perc_terras_lavoura', 'perc_terras_pastagem',
       'GEO', 'assistencia_tec', 'agricultura_fam', 'prod_escol_ens_fund',
       'geometry', 'municipio']].reset_index(drop=True)
```

```python
censo_agro
```

```python
table_name = 'censo_agro_2017'
censo_agro.to_sql(table_name, connection, if_exists='replace', index=False)
```
